```python
%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
from time import sleep

run_df = pd.read_csv('data/hkracing/runs.csv')
race_df = pd.read_csv('data/hkracing/races.csv', parse_dates=['date'])
```

```python
run_df.head()
```

```python
print(f'Average horses per race: {run_df["race_id"].value_counts().mean():.2f}')
print(f'Average odds: {run_df["win_odds"].mean()}')
```

```python
plt.figure(figsize=(10,5))
run_df.groupby('horse_age')['won'].mean().plot(color='blue', grid=True)
run_df.groupby('horse_age')['win_odds'].mean().plot(color='black', grid=True, secondary_y=True)
plt.show()
print(f'Horses per age group:\n{run_df["horse_age"].value_counts()}')
```

```python
run_df["race_id"].value_counts().value_counts()
```

```python
run_df.plot(x='horse_age', y='actual_weight', color='#c0c0c0', grid=False, kind='scatter')
run_df.groupby('horse_age')['actual_weight'].median().plot(color='red', grid=True)
run_df.groupby('horse_age')['actual_weight'].mean().plot(color='blue', grid=True)
```

```python
run_df.plot(x='actual_weight', y='declared_weight', color='#505050', grid=False, kind='scatter')
```

```python
plt.figure(figsize=(12,7))
run_df.corrwith(run_df['actual_weight']).plot(kind='bar')
plt.show()
plt.figure(figsize=(12,7))
run_df.corrwith(run_df['declared_weight']).plot(kind='bar')
plt.show()
```
